```python
faqs="""If someone who is a second and third year student
wants to do internship in MNCs and normal startups in IT sector,
then what are the topics he should prepare
for different companies like Google, Microsoft, Amazon, Paytm etc.
And the skills one should develop
like communication and all and from where he should prepare all these from,
like some YouTube channel
or some sites or resources if you can suggest.
How should he prepare for DSA
and from where suggest some good resources and what should be
the best programming language to do DSA.
And also tell about resume building
creating the LinkedIn profile
how one should update and
upgrade these.
Some companies like TCS Infosys conduct their own test
for internship, provide the data related to
these tests like when they are happening in the year,
how to apply for them
what will be the criteria and syllabus for them and how and
from where he should prepare for that.
And the last thing is what are the minor and major projects
one should do in graduation in today's
scenario, like what type of project would be beneficial for him.
Research on all these questions and give a proper response so that I can make a report on this."""
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
input_sequences=[]
for sentence in faqs.split('\n'):
 tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
 for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sequences,maxlen=max_len,padding='pre')
```

```python
padded_input_sequences
```

```python
x=padded_input_sequences[:,:-1]
```

```python
y=padded_input_sequences[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=len(tokenizer.word_index)+1)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model=Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1,output_dim=100, input_shape=(max_len-1,)))
model.add(LSTM(150))
model.add(Dense(len(tokenizer.word_index)+1,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y, epochs=100)
```

```python
import numpy as np
```

```python
text="communication"
for i in range(5):
  token_text=tokenizer.texts_to_sequences([text])[0]
  padded_token_text=pad_sequences([token_text], maxlen=max_len-1, padding='pre')
  pos=np.argmax(model.predict(padded_token_text))
  for word, index in tokenizer.word_index.items():
    if index==pos:
        text=text+' '+word
        print(text)
        break
```
